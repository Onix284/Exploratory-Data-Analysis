```python
#Load Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Convert show-id into int
df['show_id'] = df['show_id'].str.replace("s", "")
```

```python
df['show_id'] = df['show_id'].astype(int)
df.info()
```

```python
df['type'] = df['type'].map({'Movie' : 0, 'TV Show' : 1})
```

```python
df.info()
```

```python
# #For visualization
# df['title'].unique()
# df['director'].unique()

df['country'].unique()
```

```python
df.head()
```

```python
df['date_added'] = df['date_added'].str.strip()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df['Year'] =  df['date_added'].dt.year.astype('Int64')   
df['Month'] = df['date_added'].dt.month.astype('Int64')   
df['Day'] = df['date_added'].dt.day.astype('Int64')   
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('date_added', axis=1, inplace=True)
```

```python
#Duration

# Create 2 new columns: one for minutes, one for seasons
df['duration_minutes'] = df['duration'].str.extract(r'(\d+)\s*min', expand=False)
df['duration_seasons'] = df['duration'].str.extract(r'(\d+)\s*Season', expand=False)
```

```python
# Convert to numeric
df['duration_minutes'] = pd.to_numeric(df['duration_minutes'])
df['duration_seasons'] = pd.to_numeric(df['duration_seasons'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))

count = df['rating'].value_counts()

plt.pie(count,
    labels = count.index,
    autopct = '%1.1f%%',
    colors=plt.cm.Set3.colors
)

plt.title("Ratings")
plt.axis('equal')
plt.show()
```

```python
# Which are the top 10 countries producing the most content?


country_count = pd.DataFrame(df['country'].value_counts())

country_count.columns = ['Count']

```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=country_count[:10].index,
    y='Count',
    data=country_count[:10],
    palette='hls'
)

plt.title("Top Countries")
plt.show()
```
